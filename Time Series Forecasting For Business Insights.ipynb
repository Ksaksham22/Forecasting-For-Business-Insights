# Exploratory Data Analysis

```python
#Histogram
df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

#### A histographical view of overall data.

```python
#Boxplot
sns.boxplot(data=df[['Sales_Figure', 'Website_Traffic','Inventory_Level','Marketing_Spend']])
plt.show()
```

#### An overall figured  boxplot view of sales figure, website traffic, inventory level and marketing spend.

```python
#Scatter Plot
sns.pairplot(df, x_vars=['Website_Traffic','Marketing_Spend','Temperature'],y_vars = 'Sales_Figure')
plt.show()
```

#### Overview of Website traffic, Marketing spend and temperature with respect to Sales figure.

```python
plt.figure(figsize=(25, 10))
plt.plot(df['Sales_Figure'], label='Sales Figure')
plt.title('Time Series Data - Sales Figure')
plt.xlabel('Date')
plt.ylabel('Sales Figure')
plt.legend()

# Add labels on the bars
for i, sales in enumerate(df['Sales_Figure']):
    plt.text(df.index[i], sales, str(sales), ha='center', va='bottom')

plt.show()
```

#### We can clearly see from above line chart that sales figure is highest in the beginning of the month i.e from 1-5 date.

```python
# Plotting the data
fig, axs = plt.subplots(5, 1, figsize=(12, 20), sharex=True)

# Sales Figure
axs[0].plot(df['Date'], df['Sales_Figure'], marker='o', linestyle='-', color='b')
axs[0].set_title('Sales Figure Over Time')
axs[0].set_ylabel('Sales Figure')

# Website Traffic
axs[1].plot(df['Date'], df['Website_Traffic'], marker='o', linestyle='-', color='g')
axs[1].set_title('Website Traffic Over Time')
axs[1].set_ylabel('Website Traffic')

# Inventory Level
axs[2].plot(df['Date'], df['Inventory_Level'], marker='o', linestyle='-', color='r')
axs[2].set_title('Inventory Level Over Time')
axs[2].set_ylabel('Inventory Level')

# Temperature
axs[3].plot(df['Date'], df['Temperature'], marker='o', linestyle='-', color='c')
axs[3].set_title('Temperature Over Time')
axs[3].set_ylabel('Temperature (°C)')

# Marketing Spend
axs[4].plot(df['Date'], df['Marketing_Spend'], marker='o', linestyle='-', color='m')
axs[4].set_title('Marketing Spend Over Time')
axs[4].set_ylabel('Marketing Spend')
axs[4].set_xlabel('Date')

plt.tight_layout()
plt.show()
```

## Insights-

#### Sales vs. Marketing Spend: Despite a significant increase in marketing spend, sales figures do not show a corresponding increase, indicating a weak or negative correlation.

#### Website Traffic and Inventory Levels: Both website traffic and inventory levels are strongly correlated and show steady increases, suggesting that inventory levels are being adjusted based on website traffic.

#### Temperature: Temperature fluctuations seem to have a moderate correlation with sales figures, but not strongly with other variables.

## Potential Actions-

#### Marketing Strategy: Re-evaluate the marketing strategy since increased spending is not translating into higher sales.

#### Inventory Management: Maintain the strategy of adjusting inventory based on website traffic as it seems effective.
